Selection Sort

```python
import snoop
@snoop
def selection_sort(L):
  n=len(L)
  if n<1:
    return L
  for i in range(n):
    mpos=i
    for j in range(i+1,n):
      if L[j]<L[mpos]:
        mpos=j
    (L[i],L[mpos])=(L[mpos],L[i])
  return(L)
```

```python
print(selection_sort([10,4,25,1,5]))
```

Insertion Sort

```python
import snoop
@snoop
def insertion_sort(L):
  n=len(L)
  if n<1:
    return L
  for i in range(n):
    j=i
    while(j>0 and L[j]<L[j-1]):
      (L[j],L[j-1])=(L[j-1],L[j])
      j=j-1
  return L
```

```python
print(insertion_sort([10,4,25,1,5]))
```

Merge Sort

```python
import snoop
@snoop
def mergesort(a,b):
  sortedlist=[]
  x=len(a)
  y=len(b)
  i=j=0
  while(i<x and j<y):
    if a[i]<=b[j]:
      sortedlist.append(a[i])
      i=i+1
    else:
      sortedlist.append(b[j])
      j=j+1
  while i<x:
    sortedlist.append(a[i])
    i=i+1
  while j<y:
    sortedlist.append(b[j])
    j=j+1
  return sortedlist
```

```python
import snoop
@snoop
def merge(L):
  if len(L)<=1:
    return L
  n=len(L)
  left=merge(L[:n//2])
  right=merge(L[n//2:])
  B=mergesort(left,right)
  return B
```

```python
l=[2,8,9,2,6,50,44,34,37,11,29,3,2,26]
print(merge(l))
```
